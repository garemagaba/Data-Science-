```python
data_replace = {'Navi Mumbai':'Mumbai',
             'Bengaluru':'Bangalore',
             'Sharjah':'UAE',
             'Abu Dhabi':'UAE',
             'Kimberley':'Cape Town',
             'Dubai' : 'UAE'}

df_matches.replace(data_replace, inplace = True)
print('done!')
```

```python
df_matches['venue'] = df_matches['venue'].apply(lambda row: row.title().split(',')[0])  #to apply rows wise fucntion or code  #lamdba will pick each element and one by one correct it by converting it to uppercase or lower while ading spaces in between
```

```python
df_matches['venue'].value_counts().reset_index()
```

```python
data_rp = {'M.Chinnaswamy Stadium':'M Chinnaswamy Stadium',
             'Feroz Shah Kotla':'Arun Jaitley Stadium',
             'Sardar Patel Stadium':'Narendra Modi Stadium',
             'Punjab Cricket Association Is Bindra Stadium':'Punjab Cricket Association Stadium'}

df_matches['venue'].replace(data_rp,inplace = True)
print('Done')
```

```python
x = df_matches['venue'].value_counts().head(10).index
y = df_matches['venue'].value_counts().head(10).values


plt.figure(figsize = (7,5))
ax = plt.barh(x,y,color = sns.color_palette('magma',10))
plt.bar_label(ax)
plt.title('Top 10 venue by Matches Played')
plt.xticks(rotation = 90)
plt.show()
```
