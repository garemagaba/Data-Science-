```python
#Internet Required
df_titanic = sns.load_dataset('titanic')
df_titanic.sample()
```

```python
import matplotlib.pyplot as plt
sns.countplot(data = df_titanic, x = 'sex')
plt.title('Titanic Survival Analysis Bases on Gender')
plt.show()

#Julius AI
```

```python
import matplotlib.pyplot as plt
sns.countplot(data = df_titanic, x = 'sex', hue = 'alive', palette = sns.color_palette('magma',2))
plt.title('Titanic Survival Analysis Bases on Gender')
plt.show()

#Julius AI
```

```python
sns.barplot(data = df_titanic, x='class', y='fare', hue='alive', estimator='mean')
plt.show()
#black car is CI confidence interval
```

```python
sns.barplot(data = df_titanic, x ='class', y ='fare', hue='alive', errorbar=('ci',0))
plt.show()
```

```python
sns.barplot(data = df_titanic,x='class', y='fare', hue='alive', errorbar=('ci',0))
plt.grid(color = 'r', alpha = 0.1)
plt.show()
```

```python
#sns.boxplot: to detect outliners and to check data distribution
#to detect outliner
#range: min to max(19-22): 35, 28, 45(outliners: filler)

#quantile 2: 50%
#quantile : 25%
#quantile 3: 75%

#IQR = Q3 - Q1
#UW(Upper cap) = Q3 + 1.5*IQR 
#LW(Lower cap) = Q1 + 1.5*IQR
```

```python
sns.boxplot(data = df_titanic, x='class', y='fare', hue='sex')
plt.show()
```

```python
#outlier, skewness, mean value checking
```

```python
plt.figure(figsize = (10,5))
sns.boxplot(data = df_titanic, y='class', x   = 'fare' , hue = 'sex',orient = 'h')
plt.show()
#whisker: black line outside 
```

```python
# sns.violin plot

plt.figure(figsize = (10,5))
sns.violinplot(data = df_titanic, x='class', y   = 'fare' , hue = 'sex',orient = 'v')
plt.show()
```

```python
# sns.kdeplot : Kernel density estimation

sns.kdeplot(data = df_iris, x = 'sepal_length', y = 'petal_width', hue = 'species')
plt.show()
```

```python
# sns.kdeplot : Kernel density estimation

sns.kdeplot(data = df_iris, x = 'sepal_length', y = 'petal_width', hue = 'species')
sns.scatterplot(data = df_iris, x = 'sepal_length', y = 'petal_width', hue = 'species')
plt.show()
```

```python
# sns.kdeplot : Kernel density estimation

sns.kdeplot(data = df_iris, x = 'sepal_length', y = 'sepal_width', hue = 'species')
# sns.scatterplot(data = df_iris, x = 'sepal_length', y = 'sepal_width', hue = 'species')
plt.show()
```

```python
# sns.kdeplot : Kernel density estimation

sns.kdeplot(data = df_iris, x = 'sepal_length', y = 'sepal_width', hue = 'species')
sns.scatterplot(data = df_iris, x = 'sepal_length', y = 'sepal_width', hue = 'species')
plt.show()
```

```python
# sns.kdeplot : Kernel density estimation

sns.jointplot(data = df_iris, x = 'sepal_length', y = 'sepal_width', hue = 'species',) # kind : { "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" }
plt.show()
```

```python
sns.jointplot(data = df_iris, x = 'sepal_length', y = 'sepal_width',kind = 'reg') # kind : { "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" }
plt.show()
```

```python
sns.jointplot(data = df_iris, x = 'sepal_length', y = 'sepal_width',kind = 'hist', hue = 'species' ) # kind : { "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" }
plt.show()
```

```python
sns.jointplot(data = df_iris, x = 'sepal_length', y = 'sepal_width',kind = 'kde') # kind : { "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" }
plt.show()
```

```python
sns.jointplot(data = df_iris, x = 'sepal_length', y = 'sepal_width',kind = 'resid') # kind : { "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" }
plt.show()
```
