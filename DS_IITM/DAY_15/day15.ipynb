```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import *
print('Done')
```

```python
data = {'Study Hours': range(1,11),
        'Marks Obtained': [25,33,46,50,61,77,83,92,95,100]}

df = pd.DataFrame(data)
df
```

```python
plt.title('Study Versus Marks') 
plt.scatter(df['Study Hours'],df['Marks Obtained'], marker = 'x', color = 'red')
plt.xlabel('Study Hours')
plt.ylabel('Marks Obtained')
plt.show()
```

```python
# # Y = mx + c
# # M : Slope
# # C : Intercept

# Y = Marks
# x = Study_hrs

# x(1), y(1): simple Linear Regression
# x(multiple): y(1): Multiple Linear Regression
```

```python
#Import Module:
from sklearn.linear_model import LinearRegression
#Object Model Creation
model = LinearRegression()
#X, y
X = df[['Study Hours']]
y = df[['Marks Obtained']]
#Training:
model.fit(X,y)
```

```python
X = [[5.5]]
m = model.coef_
c = model.intercept_

print(m,c)
```

```python
study_hrs = [[5.5]]
marks = m*study_hrs + c

# y = mx + c
# "y =  + x"
# y = wx +  c

print('Predicted Marks:',marks[0][0])
```

```python
import warnings
warnings.filterwarnings('ignore')
study_hrs = input('Enter study hours:')

study_hrs = [[float(study_hrs)]]

ans = model.predict(study_hrs)
marks = np.round(ans[0],2)

if marks > 100:
    marks = 100

elif marks < 0:
    marks = 0

else:
    marks = marks
    
print(f'Your Predicted marks is:{marks[0]}')
```
