```python
from googlesearch import search

title = 'AI complete tutorial pdf'

data = search(title, num_results=10)

all_data = []

for i in data:
    all_data.append(i+'\n')

# Mode = x: it only create new file (fileexistserror)

file = open(f'{title}.txt', mode = 'x')
file.writelines(all_data)
file.close()
print('done')
```

```python
from googlesearch import search

title = 'Art complete tutorial pdf'

data = search(title, num_results=10)

all_data = []

for i in data:
    all_data.append(i+'\n')

# Mode = a: it only create appended file (fileexistserror)

file = open(f'{title}.txt', mode = 'a')
file.write('================================================================\n')
file.writelines(all_data)
file.close()
print('done')
```

```python
url = '''https://picsum.photos/1500/1080'''
import requests as r
n = 10
for i in range(n):
    img_data = r.get(url).content
    f = open(f'image_{i}.jpg','wb')
    f.write(img_data)
    f.close()
    from IPython import display as d
    display(d.Image(f'image_{i}.jpg', width=200, height=200))
```
